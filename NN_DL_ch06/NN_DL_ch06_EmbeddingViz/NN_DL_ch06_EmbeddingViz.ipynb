```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
MAIN_BLUE='#1A3A6E'; IDA_RED='#CD0000'; FOREST='#2E7D32'; CRIMSON='#DC3545'; AMBER='#B5853F'
```

## 1. The King-Queen Analogy
The most famous result in word embeddings:

**king - man + woman ≈ queen**


```python
# Simulated 2D embeddings showing the analogy
words = {
    'king': (2.0, 3.0), 'queen': (2.5, 3.3),
    'man': (1.0, 1.0), 'woman': (1.5, 1.3),
    'prince': (2.2, 2.5), 'princess': (2.7, 2.8),
}
fig, ax = plt.subplots(figsize=(10, 7))
for w, (x, y) in words.items():
    c = IDA_RED if w in ['king','queen'] else (FOREST if w in ['man','woman'] else MAIN_BLUE)
    ax.scatter(x, y, color=c, s=120, zorder=3)
    ax.annotate(w, (x, y), fontsize=13, fontweight='bold', color=c, xytext=(8, 8), textcoords='offset points')
# Draw the analogy arrows
ax.annotate('', xy=(2.5, 3.3), xytext=(2.0, 3.0), arrowprops=dict(arrowstyle='->', color=AMBER, lw=2.5))
ax.annotate('', xy=(1.5, 1.3), xytext=(1.0, 1.0), arrowprops=dict(arrowstyle='->', color=AMBER, lw=2.5))
ax.text(3.5, 2.0, 'king - man + woman\n≈ queen', fontsize=14, fontweight='bold',
        color=IDA_RED, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
ax.set_title('Word Embedding Analogy: Gender Direction', fontsize=14, fontweight='bold', color=MAIN_BLUE)
ax.set_facecolor('none'); fig.patch.set_alpha(0); ax.axis('off')
plt.savefig('nn_ch06_king_queen.pdf', bbox_inches='tight', dpi=300, transparent=True)
plt.savefig('nn_ch06_king_queen.png', bbox_inches='tight', dpi=150, transparent=True)
plt.show()
```

## 2. Financial Word Clusters


```python
words_f = {
    'bull': (4.5, 3.8), 'rally': (4.8, 3.5), 'gain': (4.2, 3.2), 'growth': (5.0, 3.0),
    'bear': (1.0, 1.0), 'crash': (0.8, 0.7), 'loss': (1.3, 1.3), 'decline': (1.5, 0.8),
    'inflation': (3.0, 5.0), 'rate': (3.5, 4.5), 'monetary': (2.5, 4.8),
    'bank': (6.0, 2.0), 'bond': (6.5, 2.5), 'equity': (5.5, 2.3),
}
fig, ax = plt.subplots(figsize=(10, 7))
for w, (x, y) in words_f.items():
    if w in ['bull','rally','gain','growth']: c = FOREST
    elif w in ['bear','crash','loss','decline']: c = CRIMSON
    elif w in ['inflation','rate','monetary']: c = AMBER
    else: c = MAIN_BLUE
    ax.scatter(x, y, color=c, s=80, zorder=3, alpha=0.8)
    ax.annotate(w, (x, y), fontsize=11, fontweight='bold', color=c, xytext=(5,5), textcoords='offset points')
ax.set_title('Financial Word Embeddings: Semantic Clusters', fontsize=14, fontweight='bold', color=MAIN_BLUE)
ax.set_facecolor('none'); fig.patch.set_alpha(0); ax.axis('off')
plt.savefig('nn_ch06_financial_clusters.pdf', bbox_inches='tight', dpi=300, transparent=True)
plt.savefig('nn_ch06_financial_clusters.png', bbox_inches='tight', dpi=150, transparent=True)
plt.show()
```

## 3. t-SNE Visualization


```python
clusters = {
    'Positive': (np.random.randn(25)*0.5+3, np.random.randn(25)*0.5+3, FOREST),
    'Negative': (np.random.randn(25)*0.5-2, np.random.randn(25)*0.5-2, CRIMSON),
    'Monetary': (np.random.randn(25)*0.5+0, np.random.randn(25)*0.5+4, AMBER),
    'Instruments': (np.random.randn(25)*0.5+4, np.random.randn(25)*0.5-1, MAIN_BLUE),
}
fig, ax = plt.subplots(figsize=(10, 7))
for label, (x, y, c) in clusters.items():
    ax.scatter(x, y, c=c, s=40, alpha=0.6, label=label)
ax.set_title('t-SNE of Financial Word Embeddings', fontsize=14, fontweight='bold', color=MAIN_BLUE)
ax.set_facecolor('none'); fig.patch.set_alpha(0)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), fontsize=9, ncol=4, frameon=False)
plt.savefig('nn_ch06_tsne.pdf', bbox_inches='tight', dpi=300, transparent=True)
plt.savefig('nn_ch06_tsne.png', bbox_inches='tight', dpi=150, transparent=True)
plt.show()
```
