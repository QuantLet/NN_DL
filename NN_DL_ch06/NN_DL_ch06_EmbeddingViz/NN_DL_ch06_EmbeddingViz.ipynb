## Word Embedding Visualization


```python
# Demonstrate king - man + woman = queen analogy
words = {"king": (2,3), "queen": (2.5,3.3), "man": (1,1), "woman": (1.5,1.3)}
fig, ax = plt.subplots(figsize=(8,6))
for w, (x,y) in words.items():
    ax.scatter(x, y, s=100, zorder=3)
    ax.annotate(w, (x,y), fontsize=12, fontweight="bold")
ax.set_title("Word Embedding Space", fontweight="bold")
plt.savefig("nn_ch06_word_embedding_space.pdf", bbox_inches="tight", dpi=300, transparent=True)
plt.show()
```
